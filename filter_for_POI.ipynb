```python
# Import libraries for data manipulation
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
from shapely import wkt
```

```python
fire_with_provinces =  pd.read_csv('data/fires_with_provinces.csv')
fire_with_provinces
```

```python
jasper_shapefile_url = 'model/data/Archive.zip'
jasper_gdf = gpd.read_file(f"zip://{jasper_shapefile_url}")
# Show the data in the Shapefile (States in the USA)
jasper_gdf.plot()
```

```python
fire_gdf = gpd.GeoDataFrame(
    fire_with_provinces,
    geometry=gpd.points_from_xy(fire_with_provinces['longitude'], fire_with_provinces['latitude']),
    crs=jasper_gdf.crs
)
```

```python
fire_gdf
```

```python
fire_gdf = fire_gdf.drop(columns=["index_right"])

# Perform the spatial join
data_within_shapefile = gpd.sjoin(fire_gdf, jasper_gdf, how="inner", predicate="intersects")
data_within_shapefile
```
